# Exercício 1: Seleção de Dados
Considere o seguinte DataFrame que contém dados de alunos:

```python
import pandas as pd

data = {'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
        'Idade': [20, 21, 19, 22],
        'Nota': [85, 90, 75, 88]}
df = pd.DataFrame(data)
```

```python
print(df)
```

Selecione os alunos que têm idade maior que 20 anos.

```python
# df.loc seleciona todo o dataframe e df["Idade"] seleciona a coluna idade do dataframe

maior_vinte = df.loc[df["Idade"] > 20]
print(maior_vinte)
```

Selecione os alunos que obtiveram uma nota maior ou igual a 80.

```python
nota_maior_igual = df.loc[df["Nota"]>=80]
print(nota_maior_igual)
```

# Exercício 2: Alteração de Dados
Considere o seguinte DataFrame que contém dados de vendas:

```python
data = {'Produto': ['A', 'B', 'C', 'D'],
        'Quantidade': [100, 200, 150, 300],
        'Preço_Unitário': [10, 15, 8, 20]}
df = pd.DataFrame(data)
```

```python
df
```

Adicione uma nova coluna chamada "Total_Venda" que represente o valor total de cada venda (Quantidade * Preço_Unitário).

```python
df = df.assign(Total_Venda=df["Quantidade"] * df["Preço_Unitário"])
df
```

Substitua o valor do produto 'B' na coluna 'Preço_Unitário' por 18.

```python
sub_valor = df.loc[df["Produto"]=="B", "Preço_Unitário"]=15
df
```

Considere os seguintes DataFrames que contêm informações sobre clientes e pedidos:

```python
clientes_data = {'ID': [1, 2, 3, 4],
                 'Nome': ['João', 'Maria', 'Pedro', 'Ana']}
pedidos_data = {'ID': [101, 102, 103, 104],
                'Cliente_ID': [1, 2, 3, 2],
                'Produto': ['A', 'B', 'C', 'D'],
                'Quantidade': [2, 1, 3, 2]}
clientes = pd.DataFrame(clientes_data)
pedidos = pd.DataFrame(pedidos_data)

```

```python
print(clientes)
print("---------------------")
print(pedidos)
```

Mescle os DataFrames 'clientes' e 'pedidos' com base na coluna 'Cliente_ID', criando um DataFrame final que contenha informações detalhadas sobre cada pedido, incluindo o nome do cliente.

```python
pedidos_com_nomes = pedidos.merge(clientes, left_on='Cliente_ID', right_on='ID', how='inner')

# Verifica o resultado
pedidos_com_nomes
```

# Exercício 6: Manipulação de Dados
Considere os seguintes DataFrames representando informações sobre alunos e suas notas:

```python
import pandas as pd

alunos_data = {'Aluno': ['João', 'Maria', 'Pedro', 'Ana'],
               'Idade': [20, 21, 19, 22]}
notas_data = {'Aluno': ['João', 'Maria', 'Pedro', 'Ana'],
              'Nota': [85, 90, 75, 88]}
alunos = pd.DataFrame(alunos_data)
notas = pd.DataFrame(notas_data)
```

Mescle os DataFrames 'alunos' e 'notas' com base no nome do aluno, criando um DataFrame final que contenha informações detalhadas sobre cada aluno, incluindo suas notas.

```python
df = pd.merge(alunos, notas, how="inner", on="Aluno")
df
```

Calcule a média das notas de todos os alunos.

```python
media = df["Nota"].mean()
media
```

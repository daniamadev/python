```python
# Daniela Amadeu \o/
import pandas as pd

left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": pd.Series([1, 2, 3], dtype='Int64')})

right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                     "key2": ["one", "one", "one", "two"],
                     "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
```

```python
lmerge = pd.merge(left, right, on=["key1", "key2"], how="outer")
lmerge
```

```python
linner = pd.merge(left, right, on=["key1", "key2"])
linner
```

```python
lright = pd.merge(left, right, on=["key1", "key2"], how="right")
lright
```

```python
lleft = pd.merge(left, right, on=["key1", "key2"], how="left")
lleft
```
